```python
from nltk.tokenize import RegexpTokenizer  # 정규표현식 문장에 적용
from nltk import Text
import matplotlib.pyplot as plt
```

### 텍스트 파일 읽고쓰기
#### [파일 열고 닫기]
- 파일 열기: f=open("파일경로와 이름", "파일 열기 모드")  
- 파일 닫기: f.close()
#### [읽어오는 방법]
- read(): 전체 데이터 읽어오기 
- readline(): 한줄단위 읽어오기 
- readlines(): 줄단위 요소를 리스트 형태로 읽어오기
#### [열기 모드]
- r : 읽기모드-파일을 읽기만 할때 사용(읽기전용)
- w : 쓰기모드-파일을 새로 생성해 내용을 쓸 때 사용
- a : 추가모드-기존 파일의 마지막에 내용을 추가할때 사용

```python
f=open('c:/pydata/test.txt', "r")
txt=f.read()      
print(txt)
f.close()
```

```python
plt.rcParams["figure.figsize"]=(14, 5)
retok=RegexpTokenizer("[\w]+")   # [a-zA-Z0-9_] 영문자(대소문자)와 숫자를 제외한 나머지 글자 제거
txts=Text(retok.tokenize(txt))   # Text 클래스: 문석 분석용 클래스, 토큰열 입력 생성
print(type(txts))
```

```python
plt.title('NLTK Test')
txts.plot()        # 각 단어(토큰)의 사용 빈도를 그래프로 표현(Text 클래스가 제공)
plt.show()
```

```python
txts.dispersion_plot(['you', 'for', 'set', 'free'])   # 단어가 사용된 위치를 시각화
```

```python
import nltk
nltk.download('book', quiet=True)
from nltk.book import *
```

```python
print(nltk.corpus.gutenberg.fileids())
```

```python
emma_raw = nltk.corpus.gutenberg.raw("austen-emma.txt")
print(emma_raw[:1302])
print(emma_raw[:10])
```

```python
# Text 클래스: 문석 분석용 클래스, 토큰열 입력 생성
retok=RegexpTokenizer("[\w]+")
print(Text(retok.tokenize((emma_raw[50:100]))))
```

```python
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize
```

```python
text=emma_raw[50:100]
tag_list=pos_tag(word_tokenize(text))
print(tag_list)
```

```python
nouns_list=[t[0] for t in tag_list if t[1]=='NNP' or t[1]=='NN']
print(nouns_list)
```
